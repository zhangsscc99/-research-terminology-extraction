```python
import os
filenames=os.listdir('/Users/mac/Desktop/ngram/summary/围棋')
```

```python
filenames.sort()
len(filenames)
```

```python
def change_path(files):
    for i in range(len(files)):
        files[i]='/Users/mac/Desktop/ngram/summary/围棋/'+files[i]
    return files
change_path(filenames)
```

```python
wordlist=[]
for i in range(len(filenames)):
    if filenames[i][-3:]=='txt':
        with open(filenames[i]) as f:
            wordlist.append(f.readline())

```

```python
wordlist[:50]
```

```python
wordlist
def change_wordlist(wordlist):
    for i in range(len(wordlist)):
        wordlist[i]=wordlist[i][1:-1]
    return wordlist
change_wordlist(wordlist)
```

```python
#wordlist
```

```python
#filter part
import jieba.posseg as pseg
lst=[]
for i in range(len(wordlist)):
    sentence=pseg.cut(wordlist[i])
    lst.append(sentence)
for i in range(len(lst)):
    for w in lst[i]:
        print(w.word,w.flag)
    
```

```python
filtered_lst=[]
for i in range(len(wordlist)):
    sentence=pseg.cut(wordlist[i])
    condition=True
    for w in sentence:
        if w.flag not in ['n','nz','nt','nr','ns','nw']:
            condition=False
    if condition==True:
        filtered_lst.append(wordlist[i])
    
```
