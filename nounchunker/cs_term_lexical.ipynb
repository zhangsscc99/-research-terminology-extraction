```python
import wikipedia
wikipedia.set_lang("zh")
words=wikipedia.search("围棋",results=1000)
```

```python
lst=[]
with open("wordsIT.txt",'r') as f:
    lst=f.readlines()
    
    
f.close()
```

```python
for i in range(len(lst)):
    inc=lst[i].index("\t")
    lst[i]=lst[i][:inc-1]
lst


```

```python
import copy
words=copy.deepcopy(lst)
import jieba.posseg as pseg
toshow=[]
toshow2=[]
for i in range(len(words)):
    struct=[]
    struct2=[]
    new=pseg.cut(words[i])

    for w in new:
        struct.append([w.word,w.flag])
        struct2.append(w.flag)
    toshow.append(struct)
    toshow2.append(struct2)
#toshow
toshow2
```

```python
def listToString(s):
     
    # initialize an empty string
    str1 = ""
 
    # traverse in the string
    for ele in s:
        str1 += ele
        str1+=' '
 
    # return string
    return str1
for i in range(len(toshow2)):
    toshow2[i]=listToString(toshow2[i])
toshow2
```

```python
frequency = {}

# iterating over the list
for item in toshow2:
   # checking the element in dictionary
   if item in frequency:
      # incrementing the counr
      frequency[item] += 1
   else:
      # initializing the count
      frequency[item] = 1
```
